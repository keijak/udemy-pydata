```python
import numpy as np
from pandas import Series, DataFrame
import pandas as pd
```

```python
# NFLのデータをサンプルとして使います。
import webbrowser
website = 'http://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website)
```

```python
# クリップボードから読み込むことが可能です。
nfl_frame = pd.read_clipboard()
```

```python
nfl_frame
```

```python
# 列（カラム）の名前が.columnsでわかります。
nfl_frame.columns
```

```python
nfl_frame['First NFL Season']
```

```python
# オブジェクトの属性のような書き方も可能
nfl_frame.Team
```

```python
# 特定のカラムで新しいDataFrameを作る
DataFrame(nfl_frame,columns=['Team','First Season','Total Games'])
```

```python
DataFrame(nfl_frame,columns=['Team','First Season','Total Games','Stadium'])
```

```python
# 先頭だけを表示
nfl_frame.head()
```

```python
#最後だけを表示
nfl_frame.tail()
```

```python
# indexを使って、行を取り出せる
nfl_frame.loc[3]
```

```python
# 列全体に値を代入
nfl_frame['Stadium']="Levi's Stadium" #文字列内に「'」があるので、全体は「”」で囲む
```

```python
nfl_frame
```

```python
# 長さが合っていれば、列への代入が可能。
nfl_frame["Stadium"] = np.arange(6)
nfl_frame
```

```python
# SeriesをDataFrameに追加する
stadiums = Series(["Levi's Stadium","AT&T Stadium"],index=[4,0])
```

```python
stadiums
```

```python
nfl_frame['Stadium']=stadiums
nfl_frame
```

```python
# 列を消すことも可能
del nfl_frame['Stadium']
nfl_frame
```
